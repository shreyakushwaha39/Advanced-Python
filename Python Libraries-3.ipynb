```python
#Write a program to input initial velocity and acceleration and plot the graph  depicting the equation.
#i.velocity wrt time v=u+at
#ii. Distance wrt time s=ut+0.5at**2
#iii. Distance wrt Velocity s=(v**2-u**2)/2a
import numpy as np
import pandas as pd
u=int(input("Enter initial velocity:"))
a=int(input("Enter acceleration:"))
t=np.linspace(0,10,100)
v=u+a*t
s1=u*t+0.5*a*t**2
s2=(v**2-u**2)/2*a

plt.xlabel("Velocity (v)")
plt.ylabel("Time (t)")
plt.title("Velocity wrt time")
plt.plot(v,t,color="violet",marker="*",markersize=10,markerfacecolor="blue",label="Line-1")
plt.show()

plt.xlabel("Distance (s)")
plt.ylabel("Time (t)")
plt.title("Distance wrt time")
plt.plot(s1,t,color="violet",marker="*",markersize=10,markerfacecolor="blue",label="Line-2")
plt.show()

plt.xlabel("Distance (s)")
plt.ylabel("Velocity (v)")
plt.title("Distance wrt velocity")
plt.plot(s2,t,color="violet",marker="*",markersize=10,markerfacecolor="blue",label="Line-3")
plt.show()
```

<font color='blue'><h2>Bar Graphs</h2></font>

```python
fig,ax=plt.subplots()
fruits=['apple','blueberry','cherry','orange']
counts=[40,100,30,55]
bar_labels=['red','blue','orange','green']
bar_colors=['tab:red','tab:blue','tab:orange','tab:green']
ax.bar(fruits,counts,label=bar_labels,color=bar_colors)
ax.set_ylabel("Fruit supply")
ax.set_title("Fruit supply by kind and color")
ax.legend(title='Fruit color')
plt.show()
```

```python
fig,ax=plt.subplots()
fruits=['apple','blueberry','cherry','orange']
counts=[40,100,30,55]
bar_labels=['red','blue','red','orange']
bar_colors=['tab:red','tab:blue','tab:red','tab:orange']
ax.bar(fruits,counts,label=bar_labels,color=bar_colors)
ax.set_ylabel("Fruit supply")
ax.set_title("Fruit supply by kind and color")
ax.legend(title='Fruit color')
plt.grid()
plt.show()
```

```python
species=('Adelie','Chinstrap','Gentoo')
sex_counts={'Male':np.array([73,34,61]),'Female':np.array([73,34,58])}
width=0.6

fig,ax=plt.subplots()
bottom=np.zeros(3)

for sex, sex_count in sex_counts.items():
    p=ax.bar(species,sex_count,width,label=sex,bottom=bottom)
    bottom=sex_count
    ax.bar_label(p,label_type="center")

ax.set_title("Pengins by sea")
ax.legend()
plt.show()
```
