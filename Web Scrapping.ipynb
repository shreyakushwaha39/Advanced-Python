<font color='blue'><h1>Web Scrapping:</h1></font>
* Web Scrapping is the automated process of extracting informations from websites. This information can be stored in various formats such as databases,spreadsheets or plain text. 
* Web scrapping allows larger amount of data from the internet efficiently and quickly.

<font color='red'><h2>Web Scrapping v/s Web Scrawling:</h2></font>
* Web crawling involves systematically browsing the web and downloading web pages.
* A Web Crawler is also known as **web spiders** Navigates from face to face via links, indexing contains for searh engines.

<font color='green'><h2>Uses of Web Scrapping:</h2></font>
* Market Research
* Content aggregation
* Try monitoring
* Data Analytics
* Research

<font color='green'><h2>Components of a Web Scrapper:</h2></font>
* ## HTML request:
The scraper send an HTTP request to the target website.
* ## HTML Passing:
The scraper parses the HTML to extract the desired data.
* ## Data Storage/Transform Data:
The extracted data is cleaned and transferred as needed. E.g. : csv, spreadsheet
* ## Scheduler:
Automating the scrpping process to automate data

1. <h3>Crawl:</h3>
Use libraries like 'requests' to send HTTP 'requests' and repeat web pages.

2. <h3>Parse:</h3>
Use libraries like Beautiful Soup to parse HTML content.

3. <h3>Transform:</h3>
Clean and manipulate data as required.

4. <h3>Storage</h3>
Store the data to a csv file, database or other format.

```python
#1. Import request library
#2. Send an HTTP request to get the HTML content of web page.
import requests
from bs4 import BeautifulSoup
url='https://www.niet.co.in/'
response=requests.get(url)
html_content=response.content
#3. Parse HTML content.
soup=BeautifulSoup(html_content,'html.parser')
print(soup.prettify())
```

```python
#1. Extracting Table Data from 'title' tags 
import requests
from bs4 import BeautifulSoup
url='https://www.niet.co.in/'
response=requests.get(url)
soup=BeautifulSoup(response.content,'html.parser')
print(soup.title)
print(soup.title.string)
```

```python
#2. Extracting text from "p" tags.

x=soup.find_all('p')
print(x)
for p in x:
    print(p.get_text())
```

```python
#2. Extracting text from "a" tags.
#Attributes
x=soup.find_all('a')
print(x)
for a in x:
    print(a.get_text())
```
