Why Seaborn??
Python Seaborn module serves the purpose of Data Visualization at an ease with
higher efficiency. In order to represent the variations in a huge data set,
data visualization is considered as the best way to depict and analyze the data. Seaborn stands out to have a better set of functions to carry out data 
visualization than Matplotlib in an optimized and efficient manner. 
It supports NumPy and Pandas data structure to represent the data sets.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading a data set from online repository
a=sns.load_dataset('flights')
print(a)
```

```python
#to plot lines lineplot() is used:
sns.lineplot(x='year',y='passengers',data=a)
plt.show()
```

```python
#to plot lines lineplot() is used:
sns.lineplot(x='year',y='passengers',hue='month',data=a)
plt.show()
```

use of styling a plot: style parameter must be set with following themes
    there are five different built in themes
    white,dark,whitegrid,darkgrid,ticks

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
#Bar plots: Dark Theme
plt.figure(figsize=(20,10))
sns.set(style='dark')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
#Bar plots: darkgrid
plt.figure(figsize=(20,10))
sns.set(style='darkgrid')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

Using Color palette

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='darkgrid')
sns.barplot(x='month',y='passengers',data=a,palette='rocket')
plt.show()
```

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='darkgrid')
sns.barplot(x='month',y='passengers',data=a,palette='mako')
plt.show()
```

using scatter plot

```python
b=sns.load_dataset('iris')
print(b)
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=b)
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=b)
plt.show()
```

distplot:combination of histogram and frequency plots

```python
d=sns.load_dataset('diamonds')
print(d)
```

```python
sns.distplot(d['price'],bins=10,color='red')#bins is number of bars in the plot
plt.show()
```

```python
sns.distplot(d['price'],bins=10,color='red',kde=False)#bins is number of bars in the plot
plt.show()#to remove kde plot
```

```python
sns.distplot(d['price'],bins=10,color='red',hist=False)
plt.show()# to remove hist plot
```

```python
sns.distplot(d['price'],bins=10,color='red',kde=False,vertical=True)
plt.show()#to show the plot verticaly
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=b,color='olive')
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=b,color='olive',kind='reg')
plt.show()
```

```python
x=sns.load_dataset('tips')
print(x)
```

```python
sns.catplot(x='day',y='total_bill',data=x)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=x,kind='point')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=x,kind='swarm')
plt.show()
```

```python
sns.catplot(x='day',data=x,kind='count')
plt.show()
```
